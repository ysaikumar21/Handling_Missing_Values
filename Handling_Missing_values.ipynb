**Mean_Median_Mode**

```python
def missing_data(df,var,mean,median,mode):
  df[var+'_mean']=df[var].fillna(mean)
  df[var+'_median']=df[var].fillna(median)
  df[var+'_mode']=df[var].fillna(mode)
```

```python
age_mean = df['Age'].mean()
age_median = df['Age'].median()
age_mode = df['Age'].mode()
```

```python
missing_data(df,'Age',age_mean,age_median,age_mode)
```

```python
df
```

```python
plt.figure(figsize=(5,3))


df['Age'].plot(kind='kde',color='r',label='Original')
df['Age_mean'].plot(kind='kde',color='b',label='Age_mean')
df['Age_median'].plot(kind='kde',color='g',label='Age_Median')
df['Age_mode'].plot(kind='kde',color='y',label='Age_mode')

plt.legend(loc = 0)
print(f'Original Age COlumn with NUll values : {df["Age"].std()}')
print(f'Age_mean : {df["Age_mean"].std()}')
print(f'Age_medina : {df["Age_median"].std()}')
print(f'Age_mode : {df["Age_mode"].std()}')
plt.show()
```

```python
# since mode std if close to original age so I can finalize Mode:

df = df.drop(['Age','Age_mean','Age_median'],axis=1)
df.head(7)
```

**End_of_Distribution**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_c = pd.read_csv('titanic_data.csv')
```

```python
data_c.isnull().sum()
```

```python
value = data_c['Age'].mean() + 3 * data_c['Age'].std()
value
```

```python
data_c['Age_end'] = data_c['Age'].fillna(value)
```

```python
data_c.head(7)
```

```python
plt.figure(figsize=(5,3))


data_c['Age'].plot(kind='kde',color='r',label='Original')
data_c['Age_end'].plot(kind='kde',color='b',label='Age_end')

plt.legend(loc = 0)
print(f'Original Age COlumn with NUll values : {data_c["Age"].std()}')
print(f'Age_end : {data_c["Age_end"].std()}')
plt.show()
```

**Random_sample_imputation**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_c = pd.read_csv('titanic_data.csv')
```

```python
data_c.isnull().sum()
```

```python
s = data_c['Age'].dropna().sample(data_c['Age'].isnull().sum(),random_state=42)
s
```

```python
data_c[data_c['Age'].isnull()].index
```

```python
x = data_c[data_c['Age'].isnull()].head(1)
x
```

```python
s = data_c['Age'].dropna().sample(1 , random_state = 42)
print(s)
print(s.index)
s.index = [10]
data_c['Age_c'] = s
```

```python
data_c.head(11)
```

```python
def random(df , variable):
    df[variable+"_replaced"] = df[variable].copy()
    s = df['Age'].dropna().sample(df['Age'].isnull().sum() , random_state = 42)
    s.index = df[df['Age'].isnull()].index
    df.loc[df['Age'].isnull() , 'Age_replaced'] = s

```

```python
random(data_c,'Age')
```

```python
data_c.head()
```

```python
plt.figure(figsize=(5,3))


data_c['Age'].plot(kind='kde',color='r',label='Original')
data_c['Age_replaced'].plot(kind='kde',color='b',label='Age_replaced')


plt.legend(loc = 0)
print(f'Original Age COlumn with NUll values : {data_c["Age"].std()}')
print(f'Age_replaced : {data_c["Age_replaced"].std()}')
plt.show()
```
